```python
# import pandas, librosa, os, path, re, and numpy

import pandas as pd
import librosa
import os
from pathlib import Path
import re
import numpy as np


# install moviepy, librosa, and path

'''
%%bash
pip install moviepy
pip install librosa
pip install path
'''


# import moviepy

import moviepy
from moviepy.editor import AudioFileClip
```

```python
# read in data

data = pd.read_csv("virufy-cdf-coughvid.csv")
```

```python
# remove first column; it repeats the index

data.drop(data.columns[[0]], axis = 1, inplace = True)
```

```python
# removed rows that have at least one NaN value for age, biological_sex, reported_gender, and 
# pcr_test_result_inferred

data.dropna(subset = ["age", "biological_sex", "reported_gender", "pcr_test_result_inferred"], inplace = True)
```

```python
# saving a subset of the data that will be used for the model

subset_data = data[["age", "biological_sex", "reported_gender", "pcr_test_result_inferred"]]
subset_data
```

```python
# generating .WAV files using moviepy (ONLY RUN ONCE)

count = 1
for audio in data["cough_path"]:
    moviepy.video.io.ffmpeg_tools.ffmpeg_extract_audio(audio, "WAV Audio Files/" + str(count) + ".wav")
    count += 1
```

```python
# loop through files in WAV Audio folder and append to empty list

audio = []

for file in os.listdir("WAV Audio Files/"):
    audio.append(file)
```

```python
# function to sort strings in numbers correctly

def new_sort(text): 
    reformat = lambda name: int(name) if name.isdigit() else name.lower()
    alphanumeric_key = lambda key: [reformat(i) for i in re.split('([0-9]+)', key)]
    return sorted(text, key = alphanumeric_key)
```

```python
# sort the audio files

audio = new_sort(audio)
```

```python
# ts for time series, sr for sampling rate
# extract mfcc for each file and append to empty list

mfccs = []

for file in audio:
    ts, sr = librosa.load("WAV Audio Files/" + file)
    mfccs.append(librosa.feature.mfcc(y = ts, sr = sr, n_mfcc = 40, hop_length = 500000))
```

```python
# checking that mfcc all have shape of (40, 1)

for mfcc in mfccs:
    if mfcc.shape != (40, 1):
        print(mfcc.shape)
```

```python
# make pandas dataframe from mfcc

mfcc_data = pd.DataFrame({'MFCC': mfccs})
```

```python
# reset index for mfcc dataframe to start from 1

mfcc_data = mfcc_data.reset_index(drop = True)
mfcc_data.index += 1
```

```python
# check that mfcc dataframe is reindexed correctly

mfcc_data
```

```python
# reset index for subset dataframe to start from 1

subset_data = subset_data.reset_index(drop = True)
subset_data.index += 1
```

```python
# check that subset dataframe is reindexed correctly

subset_data
```

```python
# add only column from mfcc dataframe to subset dataframe

subset_data["MFCC"] = mfcc_data["MFCC"]
subset_data
```

```python
# remove rows from pcr_test_result_inferred column that have value of "untested"

final_data = subset_data[subset_data.pcr_test_result_inferred.isin(["untested"]) == False]
final_data
```

```python
# save preprocessed dataframe as a csv file

final_data.to_csv("preprocessed_coughvid_data.csv", index = False)
```
