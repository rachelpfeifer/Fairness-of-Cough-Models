```python
# import pandas, librosa, os, path, re, and numpy

import pandas as pd
import librosa
import os
from pathlib import Path
import re
import numpy as np


# install moviepy, librosa, and path

'''
%%bash
pip install moviepy
pip install librosa
pip install path
'''


# import moviepy

import moviepy
from moviepy.editor import AudioFileClip
```

```python
# read in data

data = pd.read_csv("../Data/virufy-cdf-coughvid.csv")
```

```python
# remove first column; it repeats the index

data.drop(data.columns[[0]], axis = 1, inplace = True)
```

```python
# removed rows that have at least one NaN value for age, biological_sex, reported_gender, and 
# pcr_test_result_inferred

data.dropna(subset = ["age", "reported_gender", "pcr_test_result_inferred"], inplace = True)
```

```python
# saving a subset of the data that will be used for the model

subset_data = data[["age", "reported_gender", "pcr_test_result_inferred"]]
subset_data
```

```python
# reset indices of subset_data to start from 1

subset_data = subset_data.reset_index(drop = True)
subset_data.index += 1
subset_data
```

```python
# generating .WAV files using moviepy (ONLY RUN ONCE)

count = 1
for audio in data["cough_path"]:
    moviepy.video.io.ffmpeg_tools.ffmpeg_extract_audio(audio, "WAV Audio Files/" + str(count) + ".wav")
    count += 1
```

```python
# loop through files in WAV Audio folder and append to empty list

audio = []

for file in os.listdir("WAV Audio Files/"):
    audio.append(file)
```

```python
# function to sort strings in numbers correctly

def new_sort(text): 
    reformat = lambda name: int(name) if name.isdigit() else name.lower()
    alphanumeric_key = lambda key: [reformat(i) for i in re.split('([0-9]+)', key)]
    return sorted(text, key = alphanumeric_key)
```

```python
# sort the audio files

audio = new_sort(audio)
```

```python
# ts for time series, sr for sampling rate
# extract mfcc for each file and append to empty list

mfccs = []

for file in audio:
    ts, sr = librosa.load("WAV Audio Files/" + file)
    mfccs.append(librosa.feature.mfcc(y = ts, sr = sr, n_mfcc = 40, hop_length = 500000))
```

```python
# checking that mfcc all have shape of (40, 1)

for mfcc in mfccs:
    if mfcc.shape != (40, 1):
        print(mfcc.shape)
```

```python
# make pandas dataframe from mfcc

mfcc_data = pd.DataFrame(columns = range(40))

for mfcc in mfccs:
    mfcc_data = pd.concat([mfcc_data, pd.DataFrame(mfcc.T)], ignore_index = True)
    
mfcc_data
```

```python
# reset index for mfcc dataframe to start from 1
# rename columns from 1 to 40

mfcc_data = mfcc_data.reset_index(drop = True)
mfcc_data.index += 1
mfcc_data.columns = range(1, 41)
```

```python
# check that mfcc dataframe is reindexed correctly

mfcc_data
```

```python
# rename columns from 1 to 40

column_names = {name: str(name) for name in range(1, 41)}

mfcc_data.rename(columns = column_names, inplace = True)
```

```python
# filter out columns that have 0s in them

mfcc_data = mfcc_data[mfcc_data["3"] != 0]
```

```python
# mfcc_data dataframe

mfcc_data
```

```python
# list that stores indices of rows in mfcc_data

subset_data_indices = []

for val in mfcc_data.index:
    subset_data_indices.append(val)
```

```python
# reset index for mfcc_data dataframe to start from 1

mfcc_data = mfcc_data.reset_index(drop = True)
mfcc_data.index += 1
```

```python
# extracting indices so the corresponding rows from subset_data can be obtained

subset_data_filtered = subset_data.query('index in @subset_data_indices')
```

```python
# change age to binary labels of 1 and 0
# 35 was chosen because that was the median age of the dataset
# 1 is old (age > 35) and 0 is young (age <= 35)

subset_data_filtered.loc[subset_data_filtered["age"] <= 35, "age"] = 0
subset_data_filtered.loc[subset_data_filtered["age"] > 35, "age"] = 1
```

```python
# change age to categorical labels of old and young
# consistent with male/female in reported_gender


subset_data_filtered["age"] = subset_data_filtered["age"].map({0 : "young", 1 : "old"})
```

```python
# reset index for subset_data_filtered dataframe to start from 1
# add subset_data_filtered and mfcc_data to one dataframe
# remove rows from pcr_test_result_inferred column that have value of "untested"
# reset index for final_data dataframe to start from 1

subset_data_filtered = subset_data_filtered.reset_index(drop = True)
subset_data_filtered.index += 1

combined_data = pd.concat([subset_data_filtered, mfcc_data], axis = 1)

final_data = combined_data[combined_data.pcr_test_result_inferred.isin(["untested"]) == False]

final_data = final_data.reset_index(drop = True)
final_data.index += 1
```

```python
# save preprocessed dataframe as a csv file

final_data.to_csv("../Data/preprocessed_coughvid_data.csv", index = False)
```
