```python
# import seaborn, pandas, and numpy

import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# load in metric_results_by_gender dataframe

results_gender = pd.read_csv("metric_results_by_gender_coughvid_data.csv")
results_gender
```

```python
# define the gender metrics

gender_metrics = {
    "Accuracy": ["Female Accuracy Before", "Male Accuracy Before", "Female Accuracy After", "Male Accuracy After"],
    "Selection Rate": ["Female Selection Rate Before", "Male Selection Rate Before", "Female Selection Rate After", "Male Selection Rate After"],
    "False Negative Rate": ["Female False Negative Rate Before", "Male False Negative Rate Before", "Female False Negative Rate After", "Male False Negative Rate After"],
    "False Positive Rate": ["Female False Positive Rate Before", "Male False Positive Rate Before", "Female False Positive Rate After", "Male False Positive Rate After"],
    "True Negative Rate": ["Female True Negative Rate Before", "Male True Negative Rate Before", "Female True Negative Rate After", "Male True Negative Rate After"],
    "True Positive Rate": ["Female True Positive Rate Before", "Male True Positive Rate Before", "Female True Positive Rate After", "Male True Positive Rate After"],
    "Demographic Parity Ratio": ["Demographic Parity Ratio Gender Before", "Demographic Parity Ratio Gender After"]
}
```

```python
# set up the color palette
# red for female, blue for male

palette = {}

for metric, columns in gender_metrics.items():
    for column in columns:
        if "Female" in column:
            palette[column] = "red"
        elif "Male" in column:
            palette[column] = "blue"
        elif "Before" in column:
            palette[column] = "red"
        else:
            palette[column] = "blue"
```

```python
# make barplots with error bars showing standard error

for metric, columns in gender_metrics.items():
    plt.figure(figsize=(8, 6))
    
    results_gender_metric = results_gender[columns]
    results_gender_metric_melted = results_gender_metric.melt(var_name = "Gender", value_name = metric)
    
    sns.barplot(x = "Gender", y = metric, data = results_gender_metric_melted, estimator = np.mean, 
                errorbar = "se", capsize = 0.2, palette=palette)
    
    plt.title(f'{metric} (Before Mitigation vs. After Mitigation)')
    plt.xlabel("Gender")
    plt.ylabel(metric)
    plt.xticks(rotation = 30)
    plt.tight_layout()
    
    plt.show()
```

```python
# load in metric_results_by_age dataframe

results_age = pd.read_csv("metric_results_by_age_coughvid_data.csv")
results_age
```

```python
# set up the color palette
# green for young, blue for old

palette = {}

for metric, columns in age_metrics.items():
    for column in columns:
        if "Young" in column:
            palette[column] = "green"
        elif "Old" in column:
            palette[column] = "orange"
        elif "Before" in column:
            palette[column] = "green"
        else:
            palette[column] = "orange"
```
